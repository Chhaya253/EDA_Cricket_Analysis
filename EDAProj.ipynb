```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
# 1 Columns that should NOT be converted to numeric
exclude_cols = ['Player_Name', 'Highest_Score', 'Best_Bowling_Match']

# 2 Convert applicable object columns to numeric (using coercion for safety)
for col in df.columns:
    if df[col].dtype == 'object' and col not in exclude_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Converts invalid to NaN

# 3: Check data types now (should be mostly numeric)
print("\nUpdated Data Types:\n", df.dtypes)

# 4: Now safely apply StandardScaler on numeric features
features_to_scale = ['Runs_Scored', 'Batting_Average', 'Batting_Strike_Rate']

# Double-check: Ensure all these columns are now numeric
print("\nCheck Before Scaling:\n", df[features_to_scale].dtypes)

# Drop rows with NaN in scaling features (optional, or use imputation)
df_scaled = df.dropna(subset=features_to_scale).copy()

# Standardize
scaler = StandardScaler()
df_scaled[features_to_scale] = scaler.fit_transform(df_scaled[features_to_scale])

print("\nStandardized data preview:\n", df_scaled[features_to_scale].head())
```

 ## Part 2: Feature Engineering ##

```python
# 11. Boundary Percentage = ((4s*4 + 6s*6) / Total Runs)*100
df['Boundary_Runs'] = df['Fours'] * 4 + df['Sixes'] * 6
df['Boundary_Percentage'] = (df['Boundary_Runs'] / df['Runs_Scored']) * 100
```

```python
# 12. Batting Impact Score (custom formula)
df['Batting_Impact_Score'] = df['Runs_Scored'] * 0.5 + df['Batting_Strike_Rate'] * 0.3 + df['Batting_Average'] * 0.2
df
```

```python
# 13. All-Rounder Score: Combine normalized batting and bowling scores
df['AllRounder_Score'] = (
    df['Batting_Impact_Score'].fillna(0) +
    df['Wickets_Taken'].fillna(0) * 10 -
    df['Economy_Rate'].fillna(0)
)
df
```

```python
# 14. Categorize players based on strike rate
def classify_style(sr):
    if sr >= 140:
        return 'Aggressive'
    elif sr >= 110:
        return 'Balanced'
    else:
        return 'Defensive'

df['Batting_Style'] = df['Batting_Strike_Rate'].apply(classify_style)
df
```

```python
# 15. Year-over-Year Growth (example: Runs_Scored)
df['YoY_Runs_Growth'] = df.sort_values(['Player_Name', 'Year']).groupby('Player_Name')['Runs_Scored'].pct_change() * 100
df
```

```python
# 21. Top 5 Players by Batting Average
top_avg = df[['Player_Name', 'Batting_Average']].sort_values(by='Batting_Average', ascending=False).head(5)
print("Top 5 Players by Batting Average:\n", top_avg)
```

```python
# 22. Player with Most Matches (batting or bowling)
df['Matches_Played'] = df[['Matches_Batted', 'Matches_Bowled']].max(axis=1)
most_matches = df.loc[df['Matches_Played'].idxmax(), ['Player_Name', 'Matches_Played']]
print("\nPlayer with Most Matches:\n", most_matches)
```

```python
# 23. Players who scored more than 500 runs in any season
high_scorers = df[df['Runs_Scored'] > 500][['Player_Name', 'Year', 'Runs_Scored']]
print("\nPlayers with 500+ runs in a season:\n", high_scorers)
```

```python
# 24. Players with at least one Five-Wicket Haul
five_haul_players = df[df['Five_Wicket_Hauls'] >= 1][['Player_Name', 'Five_Wicket_Hauls']]
print("\nPlayers with ≥1 Five-Wicket Hauls:\n", five_haul_players)
```

```python
# 25. Total boundaries by each player
df['Total_Boundaries'] = df['Fours'] + df['Sixes']
boundaries = df.groupby('Player_Name')['Total_Boundaries'].sum().sort_values(ascending=False)
print("\nTotal Boundaries by Player:\n", boundaries.head(10))
```
